##### Platform Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
total = len(df['Genre'])
ax = sns.countplot(data=df, y="Genre")
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width() / total)
    x = p.get_width() + 1
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y), ha='left') 
plt.title('Platform Wise Sales')
plt.show()

```

##### from the above data, we have analyzed that distribution of games by different genres. It is clear that gamers prefer more "Action" based games from the above analysis, whereas "Puzzle" genre was least preferred.

```python
import matplotlib.pyplot as plt

names = ['Wii Sports', 'Super Mario Bros.', 'Wii Sports Resort', 'Pokemon Red/Pokemon Blue', 'Tetris', 'New Super Mario Bros.', 'Wii Play', 'New Super Mario Bros. Wii', 'Duck Hunt', 'Mario Kart Wii']
NA_sales = [41.49, 29.08, 15.75, 11.27, 23.20, 11.38, 14.03, 14.59, 26.93, 14.97]
JP_sales = [3.77, 6.81, 3.28, 10.22, 4.22, 6.50, 4.05, 6.17, 0.28, 12.88]
EU_sales = [29.02, 3.58, 12.88, 8.89, 2.26, 9.23, 9.20, 6.15, 0.28, 3.54]
Other_sales = [8.46, 0.77, 3.31, 1.00, 0.58, 6.43, 2.93, 3.60, 0.47, 2.96]
global_sales = [82.74, 40.24, 35.82, 33.00, 31.37, 30.24, 30.01, 29.02, 28.62, 28.31]

plt.figure(figsize=(12, 8))
bar_width = 0.15
index = range(len(names))

plt.bar(index, NA_sales, color='skyblue', label='NA Sales', width=bar_width)
plt.bar([i + bar_width for i in index], JP_sales, color='orange', label='JP Sales', width=bar_width)
plt.bar([i + 2*bar_width for i in index], EU_sales, color='green', label='EU Sales', width=bar_width)
plt.bar([i + 3*bar_width for i in index], Other_sales, color='red', label='Other Sales', width=bar_width)
plt.bar([i + 4*bar_width for i in index], global_sales, color='purple', label='Global Sales', width=bar_width)

plt.xlabel('Game')
plt.ylabel('Sales (Million Units)')
plt.title('Sales of Top 10 Video Games in Different Regions')
plt.xticks([i + 2*bar_width for i in index], names, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

**Wii Sports" emerges as the top-selling game globally, with significant sales in all regions, particularly in North America and Europe. 
Sales patterns vary across regions, with some games performing better in certain regions compared to others. 
For instance, "Mario Kart Wii" sees substantial sales in Japan compared to other regions. 
Overall, the chart provides insights into the popularity and regional preferences for different video games, offering valuable information for market analysis and strategic decision-making in the gaming industry.**
